<h1>Random Forest Regressor

```python
import warnings
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Load your datasets
df_allCrypto = pd.read_csv('merged_crypto_data.csv')  

# Treating Missing Value
df_allCrypto.replace(-50, np.nan, inplace=True)
df_allCrypto = df_allCrypto.dropna()
```

<h3> Main Code

```python
warnings.filterwarnings("ignore")

# Get unique crypto names
unique_crypto_names = df_allCrypto['crypto_symbol'].unique()

# Loop through each unique crypto name and filter the DataFrame
for crypto in unique_crypto_names:
    print(crypto)
    # Select data for the current crypto
    df = df_allCrypto[df_allCrypto['crypto_symbol'] == crypto]
    
    
    # Set the date column as index
    df['date'] = pd.to_datetime(df['date'])
    df.set_index('date', inplace=True)
    
    # Select the features and target variables
    features = df[['high', 'low', 'close', 'volume', 'utilization_rate', 'stable_borrow_rate']]
    
    # Display the header of the features
    print("Data Header of Features:")
    print(features.head())  # Display the first few rows, including column names
    
    # Describe data
    print("Features Data Describe:")
    print(features.describe())  
    
    target_borrowing_rate = df['borrowing_rate']
    target_lending_rate = df['lending_rate']
    
    # Normalize the features using MinMaxScaler
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_features = scaler.fit_transform(features)
    
    # Prepare the data for the Random Forest model

    X = scaled_features[:-1]  # all except the last row
    y_borrowing = target_borrowing_rate.values[1:]  # borrowing rate, shifted by one
    y_lending = target_lending_rate.values[1:]  # lending rate, shifted by one
    
    # Split the data into training and testing sets
    X_train_borrowing, X_test_borrowing, y_train_borrowing, y_test_borrowing = train_test_split(X, y_borrowing, test_size=0.2, random_state=42)
    X_train_lending, X_test_lending, y_train_lending, y_test_lending = train_test_split(X, y_lending, test_size=0.2, random_state=42)
    
    # Build the Random Forest model
    model_borrowing = RandomForestRegressor(n_estimators=100, random_state=42)
    model_lending = RandomForestRegressor(n_estimators=100, random_state=42)
    
    # Train the model for borrowing rate
    print("Train the model for borrowing rate:")
    model_borrowing.fit(X_train_borrowing, y_train_borrowing)
    
    # Train the model for lending rate
    print("Train the model for lending rate:")
    model_lending.fit(X_train_lending, y_train_lending)
    
    # Make predictions
    borrow_predicted = model_borrowing.predict(X_test_borrowing)
    lending_predicted = model_lending.predict(X_test_lending)

    
    # Evaluate the predictions
    borrow_mae = mean_absolute_error(y_test_borrowing, borrow_predicted)
    borrow_mse = mean_squared_error(y_test_borrowing, borrow_predicted)
    borrow_r2 = r2_score(y_test_borrowing, borrow_predicted)
    
    lending_mae = mean_absolute_error(y_test_lending, lending_predicted)
    lending_mse = mean_squared_error(y_test_lending, lending_predicted)
    lending_r2 = r2_score(y_test_lending, lending_predicted)
    
    # Print the evaluation metrics
    print(f'Borrowing Rate Prediction: MAE: {borrow_mae}, MSE: {borrow_mse}, R²: {borrow_r2}')
    print(f'Lending Rate Prediction: MAE: {lending_mae}, MSE: {lending_mse}, R²: {lending_r2}')
    
    # Plot the results (you can modify this to plot actual vs predicted values as needed)
    plt.figure(figsize=(14, 6))
    plt.subplot(1, 2, 1)
    plt.plot(y_test_borrowing, label='Actual Borrowing Rate')
    plt.plot(borrow_predicted, label='Predicted Borrowing Rate')
    plt.title('Borrowing Rate Prediction')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(y_test_lending, label='Actual Lending Rate')
    plt.plot(lending_predicted, label='Predicted Lending Rate')
    plt.title('Lending Rate Prediction')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
```
